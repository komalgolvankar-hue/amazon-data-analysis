etting an empty DataFrame, the issue is almost certainly that your Qty column isnt numeric or contains missing values (like ' ' or NaN), making (amazon['Qty'] > 0) false everywhere.

```python
amazon['Status'] = amazon['Status'].astype(str).str.strip()
amazon['Qty'] = pd.to_numeric(amazon['Qty'], errors='coerce')  # converts bad text to NaN
amazon['Amount'] = pd.to_numeric(amazon['Amount'], errors='coerce')

# Define valid delivery statuses

deliver_status = [
    'Shipped',
    'Shipped - Delivered to Buyer',
    'Shipped - Picked Up'
]

# Filter only delivered/valid ordred with qunatity

amazon_clean= amazon[amazon['Status'].isin(deliver_status) & (amazon['Qty'] > 0)]

# create unit price column

amazon_clean['UnitPrice']= amazon_clean['Amount']/amazon_clean['Qty']

# check results

print("cleaned data shape:", amazon_clean.shape)
print(amazon_clean.head())
```

```python
# Interpretation:
# 107,445 rows  means you have 107,445 valid sales transactions after cleaning.
#25 columns  means each sale has 25 attributes (e.g. Date, Status, Amount, Category, SKU, etc.).
```

Why:

astype(str).str.strip() removes spaces and ensures all values in Status are clean strings.

pd.to_numeric(..., errors='coerce') converts text like '5'  5 and turns bad data ('N/A') into NaN.
This makes math operations possible (e.g., dividing Amount / Qty).

Significance: You ensure data types are correct and consistent before any analysis.
Bad data types = wrong calculations or empty results.

Filtering relevant orders
Why:

Keeps only rows where the order is successfully shipped or delivered.

Removes cancelled, returned, or pending orders.

Removes rows with zero or negative quantities (which are invalid for sales).

Significance:
This isolates the valid completed sales transactions  the only ones that truly matter for revenue analysis.

Interpretation:
After this, each row represents a real product that reached the customer.
You can trust it for performance KPIs like:

Total revenue

Average selling price

Quantity sold per product

Sales trends over time


Calculating unit price

Why:
Many e-commerce datasets dont provide the per-unit selling price directly  only the total amount for that order.
Significance:
This allows you to compare pricing trends, margins, and discounts across:

Products

Categories

Time periods

Platforms (Amazon vs Flipkart)

Interpretation:
If UnitPrice varies significantly for the same SKU  discounts or dynamic pricing are happening.
If average UnitPrice drops  it might be a sign of promotions or price cuts.


Checking output
Why:
Shows how many rows were retained (after cleaning) and previews the final data.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure Date is datetime
amazon_clean['Date'] = pd.to_datetime(amazon_clean['Date'], errors='coerce')
amazon_clean = amazon_clean.dropna(subset=['Date'])

# 1 Monthly Sales Trend
monthly_sales = amazon_clean.groupby(amazon_clean['Date'].dt.to_period('M'))['Amount'].sum()

plt.figure(figsize=(12,5))
monthly_sales.plot(kind='bar')
plt.title(' Monthly Sales Trend (Amazon)')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2 Top 10 Categories by Total Revenue
top_revenue = amazon_clean.groupby('Category')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_revenue.plot(kind='bar', color='skyblue')
plt.title(' Top 10 Categories by Revenue (Amazon)')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 3 Top 10 Categories by Quantity Sold
top_qty = amazon_clean.groupby('Category')['Qty'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))

top_qty.plot(kind='bar', color='orange')
plt.title(' Top 10 Categories by Quantity Sold (Amazon)')
plt.xlabel('Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Clean Flipkart prices
flipkart['discounted_price'] = pd.to_numeric(flipkart['discounted_price'], errors='coerce')
```

```python
# Parse Flipkart category tree for main category
import ast
def parse_main(cat_str):
    try:
        parts = ast.literal_eval(cat_str)
        return parts[0].split(" >> ")[0]
    except:
        return None
flipkart['Category1'] = flipkart['product_category_tree'].apply(parse_main)
```

```python
# Convert data.csv dates
flipkart['InvoiceDate'] = pd.to_datetime(flipkart['InvoiceDate'])
```
